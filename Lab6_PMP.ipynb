```python
import numpy as np
import matplotlib.pyplot as plt
import arviz as az
from scipy.stats import gamma

a_post = 181
b_post = 11


data = gamma.rvs(a_post, scale=1/b_post, size=10000)
hdi = az.hdi(data, hdi_prob=0.94)


mode = (a_post - 1) / b_post if a_post > 1 else 0


x = np.linspace(0, 50, 1000)
y = gamma.pdf(x, a_post, scale=1/b_post)

plt.plot(x, y, label='Distribuția a posteriori')
plt.axvline(hdi[0], color='r', linestyle='--', label='HDI Inferior')
plt.axvline(hdi[1], color='g', linestyle='--', label='HDI Superior')
plt.axvline(mode, color='b', linestyle='--', label='Modul (cea mai probabilă valoare)')
plt.title('Distribuția a posteriori a ratei medii de apeluri')
plt.xlabel('Rata medie de apeluri λ')
plt.ylabel('Densitate de probabilitate')
plt.legend()
plt.grid()
plt.show()


hdi, mode
```
